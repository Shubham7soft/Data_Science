Outliers were handled using the IQR method, where values outside the range (Q1 â€“ 1.5Ã—IQR, Q3 + 1.5Ã—IQR) were removed to improve data quality.

```python
Q1 = df['AGE'].quantile(0.25)
Q3 = df['AGE'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['AGE'] >= lower_bound) & (df['AGE'] <= upper_bound)]
```

```python
df.info()
```

```python
sns.histplot(df['AGE'], kde = True)
plt.show()
```

To standardize the dataset for modeling, binary categorical features (except GENDER, AGE, and LUNG_CANCER) were recoded from 2/1 to a more conventional 1/0 format, where 1 = Yes and 0 = No, ensuring consistency for machine learning models.

```python
columns_to_convert = df.columns.difference(['GENDER', 'AGE', 'LUNG CANCER'])
df[columns_to_convert] = df[columns_to_convert].replace({2: 1, 1: 0})
```

```python
df.head()
```

```python
n_cols = 4
n_rows = (len(columns_to_convert) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 4 * n_rows))
axes = axes.flatten()


for i, col in enumerate(columns_to_convert):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(col)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')


for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='LUNG_CANCER', data=df)
plt.show()
```

We encoded the LUNG_CANCER and GENDER columns into numerical format to make them compatible with machine learning models. Specifically, we mapped "NO" to 0 and "YES" to 1 for the target variable, and "F" to 0 and "M" to 1 for gender.

```python
df['LUNG_CANCER'] = df['LUNG_CANCER'].map({'NO': 0, 'YES': 1})
df['GENDER'] = df['GENDER'].map({'F': 0, 'M': 1})
```

```python
df.head()
```

To handle class imbalance, we applied SMOTE after stratified train-test splitting. This generated synthetic samples for the minority class in the training set, creating a balanced dataset and improving model performanceâ€”especially recall on the minority class.

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

SMOTE balanced the training set, giving both classes equal samples and reducing majority-class bias.

```python
y_train_res.value_counts()
```

```python
X_train_res
```

We applied Grid Search with 5-fold Stratified Cross-Validation, optimizing for macro F1-score to handle class imbalance. Models tested included Logistic Regression, Random Forest, Gradient Boosting, KNN, SVM, and MLP. Each was trained on the SMOTE-balanced training set and evaluated on the original test set, with results compared to identify the best performer.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

y_train_res_arr = np.ravel(y_train_res)
y_test_arr = np.ravel(y_test)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = "f1_macro"

models = {
    "LogisticRegression": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", LogisticRegression(max_iter=2000, random_state=42))
        ]),
        {
            "clf__C": [0.01, 0.1, 1, 10],
            "clf__class_weight": [None, "balanced"],
            "clf__penalty": ["l2"],
            "clf__solver": ["lbfgs", "liblinear"],
        }
    ),

    "RandomForest": (
        RandomForestClassifier(random_state=42),
        {
            "n_estimators": [100, 300, 500],
            "max_depth": [None, 5, 10, 20],
            "min_samples_leaf": [1, 2, 4],
            "class_weight": [None, "balanced", "balanced_subsample"],
        }
    ),

    "GradientBoosting": (
        GradientBoostingClassifier(random_state=42),
        {
            "n_estimators": [100, 300],
            "learning_rate": [0.01, 0.1],
            "max_depth": [3, 5],
        }
    ),

    "KNN": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", KNeighborsClassifier())
        ]),
        {
            "clf__n_neighbors": [3, 5, 7, 9],
            "clf__weights": ["uniform", "distance"],
            "clf__p": [1, 2],
        }
    ),

    "SVM": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", SVC(probability=False, random_state=42))
        ]),
        {
            "clf__C": [0.1, 1, 10],
            "clf__kernel": ["linear", "rbf"],
            "clf__class_weight": [None, "balanced"],
            "clf__gamma": ["scale", "auto"],
        }
    ),

    "MLP": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", MLPClassifier(max_iter=1000, random_state=42))
        ]),
        {
            "clf__hidden_layer_sizes": [(16,), (32,), (32,16), (64,32)],
            "clf__alpha": [1e-5, 1e-4, 1e-3],
            "clf__learning_rate_init": [1e-3, 1e-2],
        }
    ),
}

all_results = []
best_models = {}

for name, (estimator, param_grid) in models.items():
    print(f"\n=== Grid Search: {name} ===")
    grid = GridSearchCV(
        estimator=estimator,
        param_grid=param_grid,
        scoring=scoring,
        cv=cv,
        n_jobs=-1,
        refit=True,
        verbose=0
    )
    grid.fit(X_train_res, y_train_res_arr)
    best_models[name] = grid.best_estimator_
    print("best parameters:", grid.best_params_)
    print(f"best {scoring}: {grid.best_score_:.4f}")

    y_pred = grid.predict(X_test)
    report_txt = classification_report(y_test_arr, y_pred, zero_division=0)
    print("\nClassification Report (test):")
    print(report_txt)

    rep = classification_report(y_test_arr, y_pred, output_dict=True, zero_division=0)
    all_results.append({
        "model": name,
        "test_accuracy": accuracy_score(y_test_arr, y_pred),
        "test_f1_macro": f1_score(y_test_arr, y_pred, average="macro"),
        "test_f1_weighted": f1_score(y_test_arr, y_pred, average="weighted"),
        "test_f1_class0": rep["0"]["f1-score"],
        "test_f1_class1": rep["1"]["f1-score"],
        "best_params": grid.best_params_,
    })

summary_df = pd.DataFrame(all_results).sort_values(by="test_f1_macro", ascending=False)
print("\n================ Summary ================")
print(summary_df.to_string(index=False))
```

# Second Modeling Attempt â€“ Optimizing for Recall (Class 1)

After evaluating models using the macro F1-score, we shifted focus to a more domain-relevant metric: recall on the positive class (cancer cases). In lung cancer detection, false negatives (missed diagnoses) are far more critical than false positives, which can be corrected through further testing.

In this phase, we:

Retained the same set of models as before

Optimized specifically for recall on class 1

Applied Stratified K-Fold Cross-Validation to ensure reliable results

This adjustment reflects a data-driven yet problem-aware approach, prioritizing the identification of as many true cancer cases as possible, even at the cost of a higher false-positive rate.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import (
    classification_report, accuracy_score, f1_score, recall_score, make_scorer
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# -----------------------------
# Datos
# -----------------------------
y_train_res_arr = np.ravel(y_train_res)
y_test_arr = np.ravel(y_test)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# -----------------------------
# MÃ©trica personalizada: recall clase positiva
# -----------------------------
scoring = make_scorer(recall_score, pos_label=1)

models = {
    "LogisticRegression": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", LogisticRegression(max_iter=2000, random_state=42))
        ]),
        {
            "clf__C": [0.01, 0.1, 1, 10],
            "clf__class_weight": [None, "balanced"],
            "clf__penalty": ["l2"],
            "clf__solver": ["lbfgs", "liblinear"],
        }
    ),

    "RandomForest": (
        RandomForestClassifier(random_state=42),
        {
            "n_estimators": [100, 300, 500],
            "max_depth": [None, 5, 10, 20],
            "min_samples_leaf": [1, 2, 4],
            "class_weight": [None, "balanced", "balanced_subsample"],
        }
    ),

    "GradientBoosting": (
        GradientBoostingClassifier(random_state=42),
        {
            "n_estimators": [100, 300],
            "learning_rate": [0.01, 0.1],
            "max_depth": [3, 5],
        }
    ),

    "KNN": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", KNeighborsClassifier())
        ]),
        {
            "clf__n_neighbors": [3, 5, 7, 9],
            "clf__weights": ["uniform", "distance"],
            "clf__p": [1, 2],
        }
    ),

    "SVM": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", SVC(probability=False, random_state=42))
        ]),
        {
            "clf__C": [0.1, 1, 10],
            "clf__kernel": ["linear", "rbf"],
            "clf__class_weight": [None, "balanced"],
            "clf__gamma": ["scale", "auto"],
        }
    ),

    "MLP": (
        Pipeline([
            ("scaler", StandardScaler()),
            ("clf", MLPClassifier(max_iter=1000, random_state=42))
        ]),
        {
            "clf__hidden_layer_sizes": [(16,), (32,), (32,16), (64,32)],
            "clf__alpha": [1e-5, 1e-4, 1e-3],
            "clf__learning_rate_init": [1e-3, 1e-2],
        }
    ),
}

all_results = []
best_models = {}

for name, (estimator, param_grid) in models.items():
    print(f"\n=== Grid Search: {name} ===")
    grid = GridSearchCV(
        estimator=estimator,
        param_grid=param_grid,
        scoring=scoring,
        cv=cv,
        n_jobs=-1,
        refit=True,
        verbose=0
    )
    grid.fit(X_train_res, y_train_res_arr)
    best_models[name] = grid.best_estimator_
    print("Best parameters:", grid.best_params_)
    print(f"Best recall (class 1): {grid.best_score_:.4f}")

    y_pred = grid.predict(X_test)
    report_txt = classification_report(y_test_arr, y_pred, zero_division=0)
    print("\nClassification Report (test):")
    print(report_txt)

    rep = classification_report(y_test_arr, y_pred, output_dict=True, zero_division=0)
    all_results.append({
        "model": name,
        "test_accuracy": accuracy_score(y_test_arr, y_pred),
        "test_f1_macro": f1_score(y_test_arr, y_pred, average="macro"),
        "test_recall_class1": rep["1"]["recall"],
        "test_precision_class1": rep["1"]["precision"],
        "test_f1_class1": rep["1"]["f1-score"],
        "best_params": grid.best_params_,
    })

summary_df = pd.DataFrame(all_results).sort_values(by="test_recall_class1", ascending=False)
print("\n================ Summary sorted by recall (class 1) ================")
print(summary_df.to_string(index=False))
```
